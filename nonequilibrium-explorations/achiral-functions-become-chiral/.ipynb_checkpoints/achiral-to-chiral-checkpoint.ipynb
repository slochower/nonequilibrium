```python
from simulation import *
from aesthetics import *
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))    
```

```python
bins = range(60)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(bins, [gaussian(i, 30, 10) for i in bins], c='b', label='Landscape 1')
ax.plot(bins, [gaussian(i, 40, 2) for i in bins], c='r', label='Landscape 2')
ax.set_ylabel('Probability')
ax.set_xlabel('Coordinate')
ax.legend(loc=2)
fetching_plot(fig, ax)
```

```python
this = simulation()
this.unbound_population = [gaussian(i, 30, 10) for i in bins]
this.bound_population = [gaussian(i, 40, 8) for i in bins]
this.simulate(pka_md_data = False)
```

```python
this.plot_energy()
this.plot_ss()
this.plot_flux()
```
